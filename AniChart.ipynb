```
#@markdown Run the top cell and then this cell once to activate interactive mode (Ctrl+F8 or run buttons on the left).

#@markdown Setting the episode slider to its maximum value also includes all shows with more than 150 episodes.

#@markdown Source material, studio and genres are pulled from AniList. Episode numbers and air dates are pulled from TMDb.org. There might still be a few mistakes from incorrectly fetched data. 

#@markdown Scores are calculated using modified Borda counts (see below).

import ipywidgets as widgets
from IPython.display import display

df['First Air Date'].replace('','00000',inplace=True)
df['Last Air Date'].replace('','00000',inplace=True)
df['First Air Date'].replace(np.nan,'00000',inplace=True)
df['Last Air Date'].replace(np.nan,'00000',inplace=True)

df['Episodes'].replace('',0,inplace=True)
df['Episodes'].replace(np.nan,0,inplace=True)

df=df.astype({'Episodes':'int','Votes':'int','Rank':'int','Esoteric Rank':'int','Episodes':'int'})

viewdf = df.dropna().drop(['AniListID','TMDbID','IMGID'],axis=1)
viewdf = viewdf.loc[viewdf['Votes']>0]

unique_studios = viewdf.groupby('Studio')['Studio'].size().index.values.tolist()
unique_source = ['All', *viewdf.groupby('Source')['Source'].size().index.values.tolist()]
if '' in unique_source:
  unique_source.remove('')

genredf = viewdf['Genres']
genredf = genredf.str.split(', ')
genredf = genredf.explode()
unique_genres = genredf.groupby(genredf).size().index.values.tolist()

def hover(hover_color='silver'):
    return dict(selector="tr:hover", props=[("background-color", "%s" % hover_color)])

def alternate_row_colors(background_color='gainsboro'):
  return dict(selector='tr:nth-child(even)',props=[("background-color", "%s" % background_color)])

styles = [hover()]

def style_df(a):
    a_styled = a.style.set_properties(**{'text-align': 'center'})#.hide_index()
    a_styled = a_styled.format("{:.1f}",subset=['Esoteric Score','Score'])
    #styles = [alternate_row_colors(),hover()]   
    styles = [hover()] 
    a_styled.set_table_styles(styles)

    return a_styled

def filt(a, Results, Year, Episodes, ExGenres, InGenres, Studio, Source, Sort, Ascending):
    
    a=viewdf[viewdf['Episodes'].notna()]

    if ExGenres != '':
        a = a[a['Genres'].str.find(ExGenres) < 0]
        
    if InGenres != '':
        a = a[a['Genres'].str.find(InGenres) > -1]

    # If slider max, include series with larger episode counts as well
    Episodes = list(Episodes)
    Episodes[1] = 1000 if Episodes[1] == 150 else Episodes[1]        
    a=a[(a['Episodes'] < Episodes[1]) & (a['Episodes'] > Episodes[0])]
    
    # Also include missing release dates
    a=a[(a['Last Air Date'].str[:4].astype(int) <= Year[1]) | (a['Last Air Date'] == '00000')]
    a=a[(a['First Air Date'].str[:4].astype(int) >= Year[0]) | (a['First Air Date'] == '00000')]
    
    if Studio != '':
        a=a[a['Studio'] == Studio]
       
    if Source not in ['All','']:
        a=a[a['Source'] == Source]
        
    if Sort != '':
        if Sort == 'Esoteric':
            a=a.sort_values(by=['Esoteric Rank'],ascending=Ascending)
        elif Sort == 'Title':
            a=a.sort_index(ascending=Ascending)
        else:
            a=a.sort_values(by=[Sort],ascending=Ascending)

    return a[:Results]

def display_df(Results, Year, Episodes, ExGenres, InGenres, Studio, Source, Sort, Ascending):

    a = filt(df, Results, Year, Episodes, ExGenres, InGenres, Studio, Source, Sort, Ascending)

    a_styled = style_df(a)
    print('Entries:', len(a))
    display(a_styled)

# Widgets

layout={'width': '350px'}

res_w = widgets.IntSlider(min=10, max=500, step=1,layout=layout,value=20,description='Results:')
year_w = widgets.IntRangeSlider(min=1970, max=2020, step=1,layout=layout,value=[1970,2020],description='Year:')
episodes_w = widgets.IntRangeSlider(min=2, max=150, step=1,layout=layout,value=[2,150],description='Episodes:')
studio_w = widgets.Combobox(options=unique_studios,layout=layout,placeholder='Studio',description='Studio:')
source_w = widgets.Dropdown(options=unique_source,layout=layout,value='All',description='Source:')
sort_w = widgets.Dropdown(options=['Rank','Esoteric','Episodes','Title','Votes','Studio','First Air Date','Last Air Date'],layout=layout,value='Rank',description='Sort:')
check_w = widgets.Checkbox(value=True,description='Ascending')

exgenres_w = widgets.Combobox(layout=layout,options=unique_genres,placeholder='Exclude Genre',description='Excl. Genre:')
ingenres_w = widgets.Combobox(layout=layout,options=unique_genres,placeholder='Include Genre',description='Incl. Genre:')

ws = [res_w,year_w,episodes_w,exgenres_w,studio_w,ingenres_w,sort_w,source_w,check_w]

out = widgets.interactive_output(display_df,{'Results':res_w,'Year':year_w,'Episodes':episodes_w,'ExGenres':exgenres_w,'InGenres':ingenres_w,'Studio':studio_w,'Ascending':check_w,'Sort':sort_w,'Source':source_w})
ui = widgets.GridBox(ws, layout=widgets.Layout(grid_template_columns="repeat(2, 400px)"))

display(ui, out)

# Adjust the output height to 2000px.
#from IPython import display
#display.Javascript("google.colab.output.setIframeHeight('2000px');")
```

```
#@title Pie Charts
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interactive_output

sns.set()

mpl.rcParams['font.size'] = 20.0
labelfontsize = 14
textprops={'fontsize': 14, 'family':'DejaVu Sans Mono'}
textcolor = 'dimgray'

# Define rectangle in axes coords
left, width = .25, .5
bottom, height = .25, .5
right = left + width
top = bottom + height

bboxprops = dict(boxstyle='round', pad=0.5, alpha=0.8, facecolor='lemonchiffon', edgecolor='darkgoldenrod')

# Slice pie in accetable sizes and put remaining pie in 'Other'
def pie_slicer(series, cm=None, thresh=0.028):
  tot = series.sum()
  other, n_slices = 0, 0
  serieslist = series.tolist()
  labels, sizes = [], []
  if 'Other' in series:
    other += series['Other']
    series.drop('Other',inplace=True)
  for i, s in enumerate(series):
    if s/tot < thresh:
      other += s
    else:
      n_slices += 1
      labels.append(series.index[i])
      sizes.append(s)
  
  if other/tot > thresh:
    labels.append('Other')
    sizes.append(other)

  colors = [plt.get_cmap('Pastel1')(i) for i in range(9)]
  [colors.append(plt.get_cmap('Pastel2')(i)) for i in range(9)]
  if cm:
    colors = [plt.get_cmap(cm)(i) for i in np.linspace(0, 1, n_slices+2)]
  return labels, sizes, colors

# Nicely format text for textbox
def box_format(series,n):
    longest = series.index.str.len().max()
    whitespaces = [''.rjust(longest-len(s)+1) if series[s] < 10 else ''.rjust(longest-len(s)) for s in series.index.values]
    t = [str(series.index.values[i])+':'+whitespaces[i]+str(series.values[i])+'\n' for i in range(len(series))]
    text = 'Top '+ str(n) +' Studios\n\n'+''.join(t)[:-1]
    return text

def pies(Top,Year,Weight):

    fig, (ax1,ax2,ax3) = plt.subplots(1, 3, figsize = (18,6))
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.4, hspace=0.6)
    
    a = viewdf[(viewdf['Rank'] <= Top[1]) & (viewdf['Rank'] >= Top[0])]

    a=a[(a['Last Air Date'].str[:4].astype(int) <= Year[1]) | (a['Last Air Date'] == '00000')]
    a=a[(a['First Air Date'].str[:4].astype(int) >= Year[0]) | (a['First Air Date'] == '00000')]

    # Source pie
    if Weight == 'None':
      g = a['Source']
      g = g.groupby(g).size().sort_values(ascending=False)
    else:
      g = a.groupby(['Source'])[Weight].sum().sort_values(ascending=False)

    labels, sizes, colors = pie_slicer(g)

    patches, texts, autotexts = ax2.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90,textprops=textprops)
    ax2.axis('equal')
    [text.set_color(textcolor) for text in texts]
    [autotext.set_color(textcolor) for autotext in autotexts]

    # Genre pie
    expl = a
    expl['Genres'] = expl['Genres'].str.split(', ')
    expl = expl.explode('Genres')
    if Weight == 'None':
      g = expl['Genres']
      g = g.groupby(g).size().sort_values(ascending=False)
    else:
      g = expl.groupby('Genres')[Weight].sum().sort_values(ascending=False)

    labels, sizes, colors = pie_slicer(g)

    patches, texts, autotexts = ax1.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90,textprops=textprops)
    ax1.axis('equal')
    [text.set_color(textcolor) for text in texts]
    [autotext.set_color(textcolor) for autotext in autotexts]

    # Top Studios
    topstudios = a.groupby('Studio')['Studio'].count()
    n = min(15,topstudios.size)
    topstudios = topstudios.sort_values(ascending=False).iloc[:n]
    text = box_format(topstudios,n)

    ax3.text(0.5*(left+right), 0.5*(bottom+top), text, family='DejaVu Sans Mono',
        horizontalalignment='center', verticalalignment='center',
        fontsize=15, color='dimgrey', transform=ax3.transAxes, bbox=bboxprops)
    ax3.axis('off')

    plt.tight_layout()
    plt.show()

# Widgets
n_titles=len(viewdf)
top_w = widgets.IntRangeSlider(min=1, max=n_titles, step=1,value=(0,n_titles),layout=layout,description='Rank:',continuous_update=False)
year_w = widgets.IntRangeSlider(min=1970, max=2020, step=1,layout=layout,value=[1970,2020],description='Year:',continuous_update=False)
check_w = widgets.Checkbox(value=False,description='Weighted:')
weight_w = widgets.Dropdown(options=['None','Score','Esoteric Score'],layout=layout,value='None',description='Weight:')

interactive_plot = interactive_output(pies, {'Top':top_w,'Year':year_w,'Weight':weight_w})
output = interactive_plot
output.layout.height = '350px'

ui = widgets.GridBox([top_w,year_w,weight_w], layout=widgets.Layout(grid_template_columns="repeat(2, 400px)"))

display(ui, output)
```


## Score Calculation

**Main Chart**

Each participant submits a list of up to 50 anime shows ranked according to personal preference. Let's say your list contains $n$ elements. Then the highest ranked show gets $n$ points, the second highest gets $n-1$ points, etc. The final ranking is determined by the total score. Shows with the same score are ranked by the number of votes.

Ideally, voters should submit fully ranked lists. However, partially ranked lists are also allowed (including unranked lists). For a list of length $n$ where the first $m$ elements are ranked, the point distribution for a show in position $p$ is as follows:

$\begin{split}
  \textrm{Points} &= n - p +1 & \quad \textrm{for} \quad p ≤ m\\
  \textrm{Points} &= \frac{1}{2} (n - m + 1) & \quad \textrm{for} \quad p > m,
\end{split}$

which results in an even point distribution for both ranked and partially ranked list.

**Esoteric Score**

The point distribution is determined by a non-linear modified Borda count that gives greater weight to shows that chart highly in lists from people who have seen many shows. Currently the formula for a show in position $m$ on a list with length $n$ is:

$\textrm{Points} = 1+(n-1) \exp(-\frac{(m-1)^2}{2 W^2})$,

where $W$ is a weight. Unranked lists currently contribute only one point per series (this might be adjusted in a later update).

Additionally, the esoteric score is multplied by an inverse popularity weighting bonus. This means that the fewer votes a series has, the greater its popularity weighting bonus becomes. Let's say the series with the most votes has $V$ votes. Then all series with $m$ votes get a multiplicative bonus of:

$\textrm{Bonus} = 1+V\exp(-\frac{(m-1)^2}{2(XV)^2})$

The bonus is then normalized by dividing it by $1+V$, so that the lowest non-zero score will always be 1. The parameter $X$ is tuned to give interesting charts. It essentially encodes what fraction of series get a sizeable bonus.

Current parameters:	$X	= 0.15$, $W	= 7$
